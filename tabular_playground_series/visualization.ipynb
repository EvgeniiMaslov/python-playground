## First look at the data

1. Missing values  - 

```python
data_path = 'data'

train = pd.read_csv(os.path.join(data_path, 'train.csv'))
test = pd.read_csv(os.path.join(data_path, 'test.csv'))

print('Train dataset shape: {}'.format(train.shape))
print('Test dataset shape: {}'.format(test.shape))

n_cat = 10
n_num = 14

categorical = ['cat'+str(i) for i in range(n_cat)]
numerical = ['cont'+str(i) for i in range(n_num)]
```

## Single variable 

```python
def make_hist(column_subset):
    subset_len = len(column_subset)
    
    fig, axs = plt.subplots(subset_len, 2, figsize=(12,22))
    fig.tight_layout()
    
    axs[0][0].set_title('Train')
    axs[0][1].set_title('Test')
    
    for index, col in enumerate(column_subset):
        sns.histplot(train[col], ax=axs[index][0])
        sns.histplot(test[col], ax=axs[index][1], color='orange')
```

```python
make_hist(categorical)
```

```python
make_hist(numerical)
```

```python
sns.displot(train['target'])
```

```python
fig, axs = plt.subplots(n_num//2, 2, figsize=(12, 18))
fig.tight_layout()

for col, ax in zip(numerical, axs.ravel()):
    sns.boxplot(data=train, x=col, ax=ax)
```

```python
for col in numerical:
    mu = train[col].mean()
    std = train[col].std()
    outliers_mask = (train[col] < (mu - 3*std)) | (train[col] > (mu + 3*std))
    print('{} contain {} outliers'.format(col, outliers_mask.sum()))
```

## Multiple variables

```python
corr_matrix = train[numerical + ['target']].corr()

_, ax = plt.subplots(1, 1, figsize=(12,12))
sns.heatmap(corr_matrix, annot=True, ax=ax)
plt.show()
```
